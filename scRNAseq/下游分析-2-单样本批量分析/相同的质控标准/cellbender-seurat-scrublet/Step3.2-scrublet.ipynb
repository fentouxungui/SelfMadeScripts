```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rc('font', size=14)
plt.rcParams['pdf.fonttype'] = 42
```

```python
for Acounts_matrix in os.listdir("./"):
    if Acounts_matrix.endswith('UMI.mtx'):
        print("***** " + Acounts_matrix + " *****")
        counts_matrix = scipy.io.mmread(Acounts_matrix).T.tocsc()
        scrub = scr.Scrublet(counts_matrix, expected_doublet_rate=0.06)
        doublet_scores, predicted_doublets = scrub.scrub_doublets(min_counts=2, 
                                                          min_cells=3, 
                                                          min_gene_variability_pctl=85, 
                                                          n_prin_comps=30)
        plt.show(scrub.plot_histogram())
        np.savetxt(Acounts_matrix.replace("UMI.mtx","scrublet.score.txt"),scrub.doublet_scores_obs_)
        
    
```
