```python
# 2. Ö±½ÓÔËÐÐ GRNBoost2 (ÕâÊÇµÚÒ»²½£¬²»Éæ¼° geneset)
adjacencies = grnboost2(expression_data=ex_matrix, tf_names=tf_names, verbose=True)
```
